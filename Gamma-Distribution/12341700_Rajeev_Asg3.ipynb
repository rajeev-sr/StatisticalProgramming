## Introduction

Marketing campaigns play a crucial role in customer acquisition and business growth. Understanding the effectiveness of a campaign requires analyzing response rates and the probability of customer conversions. In this report, we analyze the probability mass function (PMF) of the number of successful responses, evaluate how different conversion probabilities impact campaign performance, and assess the success rate distribution for different campaign sizes.

---

## Data

The study considers a marketing team reaching out to potential customers, with each customer having a fixed probability of conversion. The dataset is generated using a **binomial distribution** where:

- \( n \) is the number of contacted customers.
- \( p \) is the probability of conversion.
- $X \sim \text{Binomial}(n, p)$ represents the number of successful responses.

---

## Methodology

We use the **binomial probability mass function (PMF)** to analyze the likelihood of different numbers of successful conversions. The study includes:

1. Plotting the PMF for a campaign targeting **15 customers** with a **20% conversion probability**.
2. Examining the impact of varying probabilities (**10% to 90%**) on the PMF.
3. Investigating the success rate $( Y = \frac{X}{n} )$ for different campaign sizes (**10, 20, 50, 200**) with a **fixed conversion probability of 5%**.

---

```python
#(a)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 15  # Number of potential customers
p = 0.2  # Probability of a positive response

# Possible number of successful responses (0 to n)
k = np.arange(0, n + 1)

def plot_pmf(n, p,k):
# Calculate PMF
  pmf = binom.pmf(k, n, p)

  # Plot the PMF
  plt.figure(figsize=(12, 6))

  plt.stem(k, pmf, basefmt=" ", markerfmt='o', linefmt='-')
  plt.title("Probability Mass Function for Binomial Distribution")
  plt.xlabel("Number of Successful Responses")
  plt.ylabel("Probability")
  plt.xticks(k)
  plt.grid(axis='y', linestyle='--', alpha=0.7)


plot_pmf(n,p,k)
plt.tight_layout()
plt.show()
```

```python
#(b)
# Parameters
n = 15  # Number of potential customers
p_values = np.arange(0.1, 1.0, 0.1)  # Probabilities from 10% to 90%

# Create subplots
fig, axes = plt.subplots(3, 3, figsize=(12, 10), sharex=True, sharey=True)
axes = axes.flatten()

# Plot PMFs for different probabilities
for i, p in enumerate(p_values):
    k = np.arange(0, n + 1)
    pmf = binom.pmf(k, n, p)

    axes[i].stem(k, pmf, basefmt=" ", markerfmt='o', linefmt='-')
    axes[i].set_title(f"p = {int(p*100)}%")
    axes[i].set_xlabel("Successful Responses")
    axes[i].set_ylabel("Probability")
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

```

## Results

**Key Observations:**
- As the **conversion probability increases**, the distribution **shifts rightward**, meaning higher expected conversions.
- The **spread (variance) of responses increases**, showing greater uncertainty in outcomes.



```python
#(c)
n_sample=[10,20,50,200]
pr=0.05
fig, axes = plt.subplots(2, 2, figsize=(12, 10), sharex=True, sharey=True)
axes = axes.flatten()

for i, n in enumerate(n_sample):
    k = np.arange(0, n + 1)
    y=k/n
    pmf = binom.pmf(k, n, pr)

    axes[i].stem(y, pmf, basefmt=" ", markerfmt='o', linefmt='-')
    axes[i].set_title(f"n = {n} Customers")
    axes[i].set_xlabel("Success Rate (X/n)")
    axes[i].set_ylabel("Probability")
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()
```

## Introduction

The lifetime of a certain battery used under extremely cold conditions is modeled using a **Gamma distribution**. This report aims to calculate the expected lifetime of the battery, its variability, the median lifetime, and to generate probability density function (PDF) plots for different shape and scale parameters. These results help in understanding the expected behavior of battery lifetimes and how changing the distribution parameters impacts this behavior.

---

## Data

The data is generated using a **Gamma distribution** with the following parameters:

- **Shape parameter** \( $\alpha$ )
- **Scale parameter** \( $\theta $)

The Gamma distribution has the probability density function (PDF):

$
f(x; \alpha, \theta) = \frac{x^{\alpha - 1} e^{-x/\theta}}{\theta^{\alpha} \Gamma(\alpha)}
$

Where:
- $\alpha = 5 , \theta = 4 $ for the initial case.
- Additional cases involve changing the shape and scale parameters.

---

## Methodology

We will compute the following:
1. **Average expected lifetime** and its **variability** (standard deviation) using the Gamma distribution properties.
2. **Median lifetime**, which corresponds to the 50th percentile of the distribution.
3. **Probability Density Function (PDF) plot** for battery lifetimes in the range 0.1 to 50 hours.
4. **Repeat the calculations** for different values of the shape and scale parameters.

---

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Function to compute and display results for given shape (α) and scale (θ)
def analyze_gamma_distribution(alpha, theta):
    # (a) Calculate expected lifetime (mean) and variance
    expected_lifetime = alpha * theta
    variance = alpha * (theta ** 2)

    # (b) Find the median using the inverse CDF (percent-point function)
    median_lifetime = gamma.ppf(0.5, alpha, scale=theta)

    # Display results
    print(f"Gamma Distribution (α = {alpha}, θ = {theta}):")
    print(f"  - Expected Lifetime: {expected_lifetime} hours")
    print(f"  - Variance: {variance}")
    print(f"  - Median Lifetime: {median_lifetime} hours\n")

    # (c) Generate and plot the probability density function (PDF)
    x = np.linspace(0.1, 50, 500)
    pdf = gamma.pdf(x, alpha, scale=theta)

    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf, label=f"α={alpha}, θ={theta}")
    plt.xlabel("Battery Lifetime (hours)")
    plt.ylabel("Probability Density")
    plt.title("Gamma Distribution PDF for Battery Lifetime")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Define different parameter sets
params = [
    (5, 4),
    (10, 0.9),
    (7, 2),
    (1.5, 0.2)
]

plt.figure(figsize=(10, 6))
for alpha, theta in params:
    analyze_gamma_distribution(alpha, theta)
    print('\n')
```

## Results

### (a) Expected Lifetime and Variability (for \( \alpha = 5, \theta = 4 \))

The **mean** and **variance** for a Gamma distribution are given by:

- Mean = $\mu = \alpha \times \theta $
- Variance = $\sigma^2 = \alpha \times \theta^2 $

For the parameters $ \alpha = 5 \ and\  \theta = 4 $:

- Expected lifetime = $ 5 \times 4 = 20 $ hours
- Variability (Standard Deviation) = \( $sqrt{5 \times 4^2} = \sqrt{80} \approx 8.944 )$ hours

### (b) Median Lifetime

The **median** of a Gamma distribution can be found using the inverse cumulative distribution function (CDF). The median \( m \) satisfies:

$[
P(X \leq m) = 0.5
]$

Using numerical methods or libraries such as `scipy.stats.gamma.ppf(0.5, alpha, scale=theta)`, the median for $( \alpha = 5 )$ and $( \theta = 4 )$ is approximately **14.44 hours**.

###3. A company has developed a new type of battery whose lifetime (in hours) follows a chi-squared
distribution with 10 degrees of freedom. Using any programming language, answer the following ques-
tions:
(a) Compute the mean and standard deviation of the battery lifetime. Also, compute the median,
first quartile (Q1), and third quartile (Q3) of the lifetime.
(b) Generate a plot of the battery lifetime distribution (PDF) from 0 to 24 hours, clearly indicating
the mean, median, and quartiles on the graph. Also, make the mean, median, and quartiles on

(c) Using MGF function, generate the first, second, third moments of the chi-squared distribution

## Introduction

A company has developed a new type of battery whose lifetime (in hours) follows a **chi-squared distribution** with **10 degrees of freedom**. In this report, we analyze the statistical properties of the battery lifetime, including the mean, standard deviation, quartiles, and median. We also generate the probability density function (PDF) and compute the first three moments of the distribution using the Moment Generating Function (MGF).

---

## Data

The lifetime \( X \) of the battery follows a **chi-squared distribution** with 10 degrees of freedom:

$[
X \sim \chi^2(10)
$]

Where the **chi-squared distribution** is a special case of the **Gamma distribution** with shape parameter $( \alpha = \frac{k}{2} $) and scale parameter $( \theta = 2 $), where \( k \) is the number of degrees of freedom.

---

## Methodology

### (a) Compute the mean, standard deviation, median, first quartile (Q1), and third quartile (Q3)

For a **chi-squared distribution** with \( k = 10 \) degrees of freedom, the properties are:

- Mean: $( \mu = k $)
- Variance: $( \sigma^2 = 2k $)
- Standard Deviation: $( \sigma = \sqrt{2k} $)
  
To compute the **median**, **first quartile (Q1)**, and **third quartile (Q3)**, we will use the **percentile function** or inverse CDF.

### (b) Generate the PDF plot

We will generate a plot of the chi-squared distribution for lifetimes between 0 and 24 hours. We will mark the **mean**, **median**, **Q1**, and **Q3** on the plot to visualize the distribution.

### (c) Compute the first, second, and third moments using the Moment Generating Function (MGF)

The MGF of the chi-squared distribution with \( k \) degrees of freedom is given by:

$[
M(t) = (1 - 2t)^{-k/2} \quad \text{for} \quad t < 0.5
$]

The moments can be computed by differentiating the MGF and evaluating it at \( t = 0 \):

- First Moment (Mean): $( \mu = \frac{d}{dt} M(t) |_{t=0} )$
- Second Moment: $( \mu_2 = \frac{d^2}{dt^2} M(t) |_{t=0} $)
- Third Moment: $( \mu_3 = \frac{d^3}{dt^3} M(t) |_{t=0} )$

```python
from scipy.stats import chi2
import sympy as sp

# Given degrees of freedom
df = 10

# (a) Compute mean, standard deviation, median, Q1, and Q3
mean_lifetime = df
std_dev_lifetime = np.sqrt(2 * df)
median_lifetime = chi2.ppf(0.5, df)
q1 = chi2.ppf(0.25, df)
q3 = chi2.ppf(0.75, df)

# Display results
print(f"Chi-squared Distribution (df = {df}):")
print(f"  - Mean Lifetime: {mean_lifetime} hours")
print(f"  - Standard Deviation: {std_dev_lifetime}")
print(f"  - Median Lifetime: {median_lifetime} hours")
print(f"  - First Quartile (Q1): {q1} hours")
print(f"  - Third Quartile (Q3): {q3} hours\n")

# (b) Generate and plot the probability density function (PDF)
x = np.linspace(0, 24, 500)
pdf = chi2.pdf(x, df)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label="Chi-squared PDF", color="blue")

# Mark mean, median, Q1, and Q3 on the plot
plt.axvline(mean_lifetime, color="red", linestyle="--", label=f"Mean = {mean_lifetime:.2f}")
plt.axvline(median_lifetime, color="green", linestyle="--", label=f"Median = {median_lifetime:.2f}")
plt.axvline(q1, color="purple", linestyle="--", label=f"Q1 = {q1:.2f}")
plt.axvline(q3, color="orange", linestyle="--", label=f"Q3 = {q3:.2f}")

plt.xlabel("Battery Lifetime (hours)")
plt.ylabel("Probability Density")
plt.title("Chi-squared Distribution of Battery Lifetime")
plt.legend()
plt.grid(True)
plt.show()

# (c) Compute first, second, and third moments using the Moment Generating Function (MGF)
def chi_squared_mgf(t, df):
      return (1 - 2*t) ** (-df / 2)

t=sp.symbols('t')
t_val=0
first_moment = sp.diff(chi_squared_mgf(t,df),t) # Mean (expected value)
second_moment = sp.diff(first_moment,t)
third_moment = sp.diff(second_moment,t)
first_moment=first_moment.subs(t,t_val)
second_moment=second_moment.subs(t,t_val)
third_moment=third_moment.subs(t,t_val)

print()
# Display moment values
print(f"First Moment (E[X]): {first_moment}")
print(f"Second Moment (E[X^2]): {second_moment}")
print(f"Third Moment (E[X^3]): {third_moment}")
```

## Results

### (a) Mean, Standard Deviation, Median, and Quartiles

For \( k = 10 \):

- **Mean**: $( \mu = 10 $) hours
- **Standard Deviation**: $( \sigma = \sqrt{2 \times 10} = \sqrt{20} \approx 4.472 $) hours
- **Median**: $( \text{Median} \approx 8.93 $) hours (computed using inverse CDF)
- **First Quartile (Q1)**: $( Q1 \approx 5.19 $) hours
- **Third Quartile (Q3)**: $( Q3 \approx 14.71 $) hours

### (b) Probability Density Function (PDF) Plot

### (c) First Three Moments using MGF

The Moment Generating Function (MGF) for the chi-squared distribution with 10 degrees of freedom is:

$[
M(t) = (1 - 2t)^{-5}
]$

Using this, we calculate the first three moments:

1. **First Moment (Mean)**:  
   $
   \mu = \frac{d}{dt} \left[ (1 - 2t)^{-5} \right] |_{t=0} = 10
   $
   
2. **Second Moment**:  
   $
   \mu_2 = \frac{d^2}{dt^2} \left[ (1 - 2t)^{-5} \right] |_{t=0} = 20
   $

3. **Third Moment**:  
   $
   \mu_3 = \frac{d^3}{dt^3} \left[ (1 - 2t)^{-5} \right] |_{t=0} = 1680
   $

These values confirm the distribution's characteristics.
