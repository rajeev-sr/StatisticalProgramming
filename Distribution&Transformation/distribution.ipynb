# Question 1
Imagine you are studying two distinct process that generate random numbers between 0 and 1, modeled
as continuous random variables X with different distributions:

a. The first process generates numbers, following an exponential distribution with λ = 2.

b. The second process generates numbers uniformly distributed between 0 and 1.

You collect n random numbers from each process and define a new random variable, Y , given by Y = FX (x) where FX (x) represents the Cumulative Distribution Function or CDF of X. Derive the
Probability Distribution Function (PDF) of Y for both processes and Observe the histogram of generated n numbers for both processes for different values of n. Write down your observations.

---
---

## Introduction
This report analyzes two distinct processes that generate random numbers between 0 and 1. The processes are modeled as continuous random variables `X` with the following distributions:
1. The first process follows an **exponential distribution** with rate parameter `λ = 2`.
2. The second process follows a **uniform distribution** between 0 and 1.

We define a new random variable, `Y = F_X(x)`, where `F_X(x)` represents the Cumulative Distribution Function (CDF) of `X`. The goal is to derive the Probability Distribution Function (PDF) of `Y` for both processes, simulate the data for different values of `n`, and observe the corresponding histograms.

---

## Data
- **Process 1 (Exponential Distribution)**:
  - PDF: `f_X(x) = 2 * exp(-2 * x)` for `x >= 0`
  - CDF: `F_X(x) = 1 - exp(-2 * x)` for `x >= 0`

- **Process 2 (Uniform Distribution)**:
  - PDF: `f_X(x) = 1` for `0 <= x <= 1`
  - CDF: `F_X(x) = x` for `0 <= x <= 1`

---

## Methodology
1. **Derivation of PDFs for Y**:
   - For **Process 1**:
     - Let `Y = F_X(x) = 1 - exp(-2 * x)` with inverse `x = -0.5 * ln(1 - Y)`.
     - Using the transformation rule for PDFs: `f_Y(y) = f_X(x) * |dx/dy|`, we get:
       ```
       f_Y(y) = 1 for 0 <= y <= 1
       ```
     - This implies `Y` follows a **uniform distribution**.

   - For **Process 2**:
     - Since `F_X(x) = x`, `Y = F_X(x)` directly results in:
       ```
       f_Y(y) = 1 for 0 <= y <= 1
       ```
     - Here, `Y` also follows a **uniform distribution**.

2. **Simulation**:
   - Generate `n` random samples for both processes.
   - Compute `Y = F_X(x)` using the respective CDFs.
   - Plot histograms of the transformed variable `Y` for varying values of `n` (e.g., `n = 10, 100, 1000`).

3. **Visualization**:
   - Compare histograms for both processes to confirm the derived PDFs.

---

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon, uniform


n_val=[1000,10000,100000,1000000,10000000]
lmda=2
for n in n_val:
  # Process 1: Exponential Distribution
  exp_x=np.random.exponential(scale=1/lmda,size=n)
  exp_y=expon.cdf(exp_x,scale=1/lmda)

  # Process 2: Uniform Distribution
  unif_x=np.random.uniform(0,1,size=n)
  unif_y=uniform.cdf(unif_x)

  plt.figure(figsize=(12,6))

  # Exponential Process Histogram
  plt.subplot(1,2,1)
  plt.hist(exp_y, bins=25,color='red',alpha=0.7,edgecolor='black',label='Process 1 (Exponential lamda=2)')
  plt.title(f'Histogram of Y (Process 1 Exponential) for n={n}')
  plt.xlabel('Y')
  plt.ylabel('Frequency')
  plt.legend()

  # Uniform Process Histogram
  plt.subplot(1,2,2)
  plt.hist(unif_y, bins=25,color='green',alpha=0.7,edgecolor='black',label='Process 2 (Uniform)')
  plt.title(f'Histogram of Y (Process 2 Uniform) for n={n}')
  plt.xlabel('Y')
  plt.ylabel('Frequency')
  plt.legend()

  plt.tight_layout()
  plt.show()

````

## Results
1. **Process 1 (Exponential Distribution)**:
   - The histogram of `Y` aligns with a uniform distribution for all values of `n`.
   - As `n` increases, the histogram becomes smoother and closely resembles the theoretical uniform PDF.

2. **Process 2 (Uniform Distribution)**:
   - The histogram of `Y` directly represents a uniform distribution, as expected.
   - Increasing `n` similarly leads to smoother and more accurate representations of the theoretical PDF.

---

## Discussion
### Observations:
1. Both processes result in `Y` being uniformly distributed between 0 and 1.
2. This outcome aligns with theoretical expectations since `F_X(x)` maps any continuous distribution to a uniform distribution when applied as a transformation.

### Key Insights:
- The transformation using the CDF standardizes the distribution, making it easier to compare across different processes.
- The histogram's smoothness improves with larger sample sizes (`n`), highlighting the importance of sufficient data for accurate modeling.

---

## Conclusion
- The derived PDFs and simulations confirm that for both processes, the transformed variable `Y` follows a uniform distribution between 0 and 1.
- The histograms validate the theoretical derivations, with increasing sample size improving accuracy.
- This method can be extended to analyze other distributions, showcasing the utility of the CDF transformation in statistical modeling and comparison.

# Question 2
Imagine you are an archivist analyzing a dusty old text file from a forgotten library. Your task is to
uncover the hidden patterns in the text by doing the following:

a. Count the Words: Create a histogram of how often each word appears.

b. Focus on the Key Players: Identify the top 30 most frequent words from your list.

Once you have this list of important words, calculate the Cumulative Distribution Function (CDF) for their frequencies. Use this CDF as a transformation function to remap the word frequencies. What new insights can you uncover about the text after applying this transformation?

---
---


## Introduction
The task involves analyzing a text file to uncover patterns in word frequencies. The steps include:
1. Counting the frequency of each word in the text.
2. Identifying the top 30 most frequent words.
3. Calculating the Cumulative Distribution Function (CDF) for the frequencies of these top words.
4. Using the CDF as a transformation function to remap the word frequencies.

The goal is to analyze and gain insights into the text's structure and the prominence of certain words.

---

## Data
The analysis is based on a text file containing a large corpus of words. The extracted word frequencies revealed the following:
- The most frequent word is "the," with a frequency of 1804.
- The top 30 words and their frequencies are provided, starting with:
  - "the" (1804), "of" (804), "and" (681), "to" (614), etc.

This list forms the foundation for further analysis.

---

## Methodology
1. **Word Count**:
   - Counted all words in the text and stored their frequencies in a dictionary.
   - Sorted the dictionary to identify the top 30 words.

2. **CDF Calculation**:
   - Calculated the cumulative sum of the top 30 word frequencies.
   - Normalized the cumulative sum by dividing it by the total frequency of the top 30 words.

3. **Frequency Transformation**:
   - Applied the CDF as a transformation to remap word frequencies.
   - The transformed values represent the relative importance of words based on their cumulative contribution to the total frequency.

---


## Discussion
### Insights:
1. The most frequent words are common articles and prepositions, such as "the," "of," and "and."
2. The CDF shows that the first few words (e.g., the top 10) account for a large cumulative portion of the total frequency, indicating their dominance in the text.
3. Applying the CDF transformation highlights how quickly the contribution of each additional word diminishes, reinforcing the power-law nature of word frequencies in language.

### Applications:
- The insights can be applied to prioritize certain words for text summarization, sentiment analysis, or keyword extraction.
- The dominance of specific words could also reflect the style or purpose of the text, such as formal writing or technical documentation.


````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

filepath="text_file.txt"
f=open(filepath,'r')

word_freq={}

for line in f:
  for word in line.split():
    if word.isalpha():
      if word.lower() in word_freq:
        word_freq[word.lower()]+=1
      else:
        word_freq[word.lower()]=1

````

````python
sorted_word_freq=sorted(word_freq.items(),key=lambda x:x[1],reverse=True)
df=pd.DataFrame(sorted_word_freq,columns=['word','frequency']);
df=df.loc[:29]
plt.figure(figsize=(12, 6))
plt.bar(df['word'],df['frequency'],color='blue',alpha=0.7,edgecolor='black')
plt.xticks(rotation=90)
plt.title('bargrapgh of Word Frequencies')
plt.xlabel('Word Frequency')
plt.ylabel('Frequency')
plt.show()
# print(df.loc[:29])
````

````python

print("top 30 words : ")
#most frequent words
print(df)

most_freq=list(map(lambda x:x[1],sorted_word_freq[:30]))
````

````python
total = sum(most_freq)
cumulative = np.cumsum(most_freq)
cdf=cumulative/total
print(cdf)
````

## Results
- The top 30 words and their corresponding frequencies were extracted, with the cumulative frequencies plotted as a CDF.
- The CDF values for the top 30 words range from `0.209` to `1.0`, as shown in the image:
  - For example:
    - The CDF for "the" is approximately `0.209`.
    - The CDF for "of" is `0.302`, and so on.

The transformation effectively emphasizes the relative contribution of each word to the total frequency.

---

````python
most_freq_words=list(map(lambda x:x[0],sorted_word_freq[:30]))
def remap_frequencies(words, cdf):
    remaped_freq = {}
    i=0;
    for word in words:
        remaped_freq[word] = cdf[i]
        i=i+1
    return remaped_freq

remaped_freq=remap_frequencies(most_freq_words,cdf)
df['remaped_frequency']=remaped_freq.values()
print(df)
````

## Conclusion
The analysis revealed that:
- The word "the" is the most dominant, followed by other common function words.
- The CDF transformation highlights the unequal contribution of words, with the top 10 words accounting for a significant portion of the total frequency.
- This method can be extended to analyze other texts, revealing structural or stylistic patterns and informing tasks like language modeling or content analysis.

The process underscores the utility of statistical tools like histograms and CDFs in extracting meaningful patterns from text data.

Imagine you are working with a random number U , which is drawn from a uniform distribution between
0 and 1. You have a tool that allows you to transform this number by using the inverse CDF of a
distribution X. By applying this transformation, you create a new number Y .
Now, your task is to determine what kind of random variable Y becomes after the transformation.
Consider the following two cases:

a. When X ∼ Exponential

b. When the PDF of distribution is as follows:

Screenshot from 2025-01-23 13-24-54.png

---
---

## Data

### Case a: Exponential Distribution
- The random variable `X` is exponentially distributed with rate parameter `λ > 0`.
- The **PDF** (Probability Density Function) for `X` is:

                        f_X(x) = λ * exp(-λ * x), for x >= 0

- The **CDF** (Cumulative Distribution Function) for `X` is:
                        F_X(x) = 1 - exp(-λ * x), for x >= 0
  

### Case b: Custom Distribution
- The **PDF** for `X` is defined as:
                       f_X(x) = x for 0 <= x <= 1,
                              = 2 - x for 1 < x <= 2,
                              = 0 ,otherwise

- The **CDF** for `X` is:
                        F_X(x) = 0 for x < 0,
                              = (x^2) / 2 for 0 <= x <= 1,
                              = 2x - (x^2) / 2 - 1 for 1 < x <= 2,
                              = 1 for x > 2

- The **Inverse CDF** for `X` is computed as:
- `x = sqrt(2u)` for `u` in `[0, 0.5]`
- `x = 2 - sqrt(2(1 - u))` for `u` in `(0.5, 1]`

---

## Methodology

### Random Variable Transformation
1. Generate `U ~ Uniform(0, 1)`.
2. Transform `U` using the inverse CDF of the respective distribution:
 - For the exponential case:
   ```
   Y = -ln(1 - U) / λ
   ```
 - For the custom distribution:
   ```
   Y = sqrt(2U),                for 0 <= U <= 0.5,
       = 2 - sqrt(2(1 - U)),      for 0.5 < U <= 1
   ```

### Simulation and Visualization
- Generate [100,500,1000,10000,100000] samples of `U`.
- Transform `U` using the above formulas for both cases.
- Plot histograms of the transformed variable `Y` and overlay the theoretical PDF.

### Validation
- Ensure that the histogram matches the theoretical PDF by comparing both visually and statistically.

---

## Discussion

### Transformation via Inverse CDF
- The inverse CDF method effectively maps the uniform random variable `U` to the desired target distribution `X`.
- For the exponential distribution, the memoryless property of the exponential random variable is preserved.
- For the custom distribution, the transformation captures the triangular density's linear rise and fall.

### Numerical Accuracy
- Accuracy is contingent on numerical computation of logarithm (`ln`) and square root (`sqrt`), particularly for edge cases like `U` near `0` or `1`.

---


````python
import numpy as np
import matplotlib.pyplot as plt

# Case a: Exponential Distribution
def inverse_cdf_exponential(u, lambd):
    return -np.log(1 - u) / lambd

# Case b: Custom PDF (Triangular)
def inverse_cdf_custom(u):
    result = np.zeros_like(u)
    # print(len(result))
    # For 0 <= u <= 0.5
    interval1 = (u >= 0) & (u <= 0.5)
    result[interval1] = np.sqrt(2 * u[interval1])
    # For 0.5 < u <= 1
    interval2 = (u > 0.5) & (u <= 1)
    result[interval2] = 2 - np.sqrt(2 * (1 - u[interval2]))
    return result

# Generate random uniform samples
n = [100,500,1000,10000,100000]

for n_samples in n :

  U = np.random.uniform(0, 1, n_samples)

  # Exponential Transformation
  lambda_param = 1.0  # Rate parameter for Exponential
  Y_exp = inverse_cdf_exponential(U, lambda_param)

  # Custom Transformation
  Y_custom = inverse_cdf_custom(U)

  # Plot the results
  plt.figure(figsize=(12, 6))

  # Exponential Distribution
  plt.subplot(1, 2, 1)
  plt.hist(Y_exp, bins=50, alpha=0.7, color='blue', label='Transformed Samples')
  plt.title(f"Exponential Distribution for n={n_samples}")
  plt.xlabel("Y")
  plt.ylabel("Density")
  plt.legend()

  # Custom Distribution
  plt.subplot(1, 2, 2)
  plt.hist(Y_custom, bins=50, alpha=0.7, color='green', label='Transformed Samples')
  plt.title(f"Custom Triangular Distribution  for n={n_samples}")
  plt.xlabel("Y")
  plt.ylabel("Density")
  plt.legend()
  plt.tight_layout()
  plt.show()
````

## Results

### Exponential Distribution
- The transformed samples `Y` closely follow the exponential distribution with PDF:
                          f_X(x) = λ * exp(-λ * x)

- For example, with `λ = 1`, the histogram of `Y` aligns well with the expected exponential curve.

### Custom Distribution
- The transformed samples `Y` match the triangular PDF:
                          f_X(x) = x for 0 <= x <= 1,
                                 = 2 - x for 1 < x <= 2

- The histogram increases linearly from `0` to `1` and decreases linearly from `1` to `2`, matching the triangular shape.


---

## Conclusion
- The study validates the inverse CDF method as a robust technique for transforming uniform random variables into any desired distribution.
- The transformed variable `Y` matches the desired target distributions in both cases:
- An exponential distribution for `X ~ Exponential(λ)`.
- A triangular distribution for the custom PDF.
- This approach can be applied to simulate data from arbitrary distributions, provided the inverse CDF is known or can be computed.
