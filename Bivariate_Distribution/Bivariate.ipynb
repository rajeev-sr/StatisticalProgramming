# Ques 1) Bivariate Normal Distribution Probabilities

## Introduction
This aims to compute various probabilities related to a bivariate normal distribution using given statistical parameters. We will analyze both marginal and conditional probabilities to understand the distribution characteristics.

## Data
The given parameters for the bivariate normal distribution are:
- Mean of X muX = 3
- Mean of Y mu_Y = 1
- Standard deviation of X sigma_X = 4
- Standard deviation of Y sigma_Y = 5
- Correlation coefficient rho_{XY} = 3/5

## Methodology
Using the `scipy.stats` library, we compute the following probabilities:
1.  P(3 < Y < 8)
2.  P(3 < Y < 8 | X = 7)
3.  P(-3 < X < 3)
4.  P(-3 < X < 3 | Y = -4)

We utilize the cumulative distribution function (CDF) of the normal distribution to compute these probabilities, leveraging conditional expectation formulas where necessary.


```python
# 1. Let X and Y have a bivariate normal distribution with parameters:
# µx = 3,
# µy = 1,
# σx2 = 16,
# σy2 = 25,
# ρxy =
# 3
# 5
# where µx and µy represents mean of X and Y , σx2 and σy2 represents variance of X and Y and ρxy
# represents correlation coefficient between of X and Y .
# Determine the following probabilities:
# (a) P (3 < Y < 8).
# (b) P (3 < Y < 8 | X = 7)
# (c) P (−3 < X < 3)
# (d) P (−3 < X < 3 | Y = −4
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.stats as stats
mux=int(input("mux value:"))
muy=int(input("muy value:"))
sigma_x_2=int(input("varx value:"))
sigma_x=math.sqrt(sigma_x_2)
sigma_y_2=int(input("vary value:"))
sigma_y=math.sqrt(sigma_y_2)
rho_xy=float(input("correlation bw x and y:"))
prob_a = stats.norm.cdf(8, loc=muy, scale=sigma_y) - stats.norm.cdf(3, loc=muy, scale=sigma_y)

# (b) P(3 < Y < 8 | X = 7)
mu_y_given_x7 = muy + rho_xy * (sigma_y / sigma_x) * (7 - mux)
sigma_y_given_x7 = sigma_y * (1 - rho_xy**2)**0.5
prob_b = stats.norm.cdf(8, loc=mu_y_given_x7, scale=sigma_y_given_x7) - stats.norm.cdf(3, loc=mu_y_given_x7, scale=sigma_y_given_x7)

#     # (c) P(-3 < X < 3)
prob_c = stats.norm.cdf(3, loc=mux, scale=sigma_x) - stats.norm.cdf(-3, loc=mux, scale=sigma_x)

#     # (d) P(-3 < X < 3 | Y = -4)
mu_x_given_y_neg4 = mux + rho_xy * (sigma_x / sigma_y) * (4 - muy)
sigma_x_given_y_neg4 = sigma_x * (1 - rho_xy**2)**0.5
prob_d = stats.norm.cdf(3, loc=mu_x_given_y_neg4, scale=sigma_x_given_y_neg4) - stats.norm.cdf(-3, loc=mu_x_given_y_neg4, scale=sigma_x_given_y_neg4)

# Results
print("Results:")
print(f"(a) P(3 < Y < 8) = {prob_a:.4f}")
print(f"(b) P(3 < Y < 8 | X = 7) = {prob_b:.4f}")
print(f"(c) P(-3 < X < 3) = {prob_c:.4f}")
print(f"(d) P(-3 < X < 3 | Y = 4) = {prob_d:.4f}")
# find probabilities
```

# **Analysis of Chi-Squared Distribution in Multivariate Normal Data**

## **Introduction**
This report investigates the distribution of the quadratic form \((x - \mu)^T \Sigma^{-1} (x - \mu)\) in a multivariate normal setting. This expression follows a Chi-squared distribution with degrees of freedom equal to the dimensionality of the dataset. We analyze this property across different dimensions and sample sizes.

## **Data**
We generate synthetic datasets with varying dimensions:

- **Dataset 1:** 2-dimensional data, 500 samples
- **Dataset 2:** 3-dimensional data, 1000 samples
- **Dataset 3:** 4-dimensional data, 10,000 samples

The mean vectors (\(\mu\)) and covariance matrices (\(\Sigma\)) for each dataset are predefined.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import multivariate_normal

def generate_samples(mu, sigma, P):
    # a) Generates samples
    return multivariate_normal.rvs(mean=mu, cov=sigma, size=P)


def calculate_y(X, mu, sigma_inv):
    #Calculates Y
    return np.diag((X - mu) @ sigma_inv @ (X - mu).T)


def calculate_probability(mu, sigma_inv, c):
  # For an n-dimensional multivariate normal distribution, (x - µ)T Σ−1 (x - µ) follows a Chi-squared distribution with n degrees of freedom.
  n = len(mu) # Number of dimensions
  return stats.chi2.cdf(c**2, n)


# Example usage
n1 = 2  # Dimensionality
P1 = 1000 # Number of samples
mu1 = np.array([0, 0])  # Mean vector
Sigma1 = np.array([[1, 0.5], [0.5, 1]])  # Covariance matrix

# Define parameters
n2 = 3  # Dimensionality
P2 = 1000  # Number of samples
mu2 = np.array([3, 2, 3])
Sigma2 = np.array([[2, 0.5, 0.3], [0.5, 1, 0.2], [0.3, 0.2, 1]])  # Example covariance matrix

n3 = 4  # Dimensionality
P3 = 10000  # Number of samples
mu3 = np.array([3, 2, 3,1])
Sigma3 = np.array([[2, 1, 2,0.5], [1, 3, 0.4,2], [2, 0.4, 4,1],[0.5,2,1,5]])  # Example covariance matrix

c = 2

def mul(n,mu,sigma,c,P):
    # Check if sigma is positive semidefinite
    sigma_inv = np.linalg.inv(sigma)


    # (a) Generate samples
    X = generate_samples(mu, sigma, P)

    # (b) Calculate Y values
    Y = calculate_y(X, mu, sigma_inv)
    # Observe the distribution of Y (you might want to plot a histogram)01
    plt.hist(Y, bins=30,color='red',edgecolor='black',alpha=0.6)
    plt.title(f"Distribution of Y for n={n} and P={P}")
    plt.xlabel("Y")
    plt.ylabel("Frequency")
    plt.show()

    # (c) Calculate Probability
    prob = calculate_probability(mu, sigma_inv, c)
    print(f"Probability that (x − µ)T Σ−1 (x − µ) ≤ {c**2}: {prob}")
    print()

mul(n1,mu1,Sigma1,c,P1)
mul(n2,mu2,Sigma2,c,P2)
mul(n3,mu3,Sigma3,c,P3)
```

## **Methodology**
1. **Sample Generation**: We generate random samples from a multivariate normal distribution using the specified \(\mu\) and \(\Sigma\).
2. **Quadratic Form Calculation**: We compute the quadratic form \(Y = (x - \mu)^T \Sigma^{-1} (x - \mu)\) for each sample.
3. **Distribution Analysis**: We visualize the distribution of \(Y\) using histograms.
4. **Probability Calculation**: We compute the probability that \(Y \leq c^2\) using the Chi-squared cumulative distribution function (CDF), where \(c = 2\).

## **Results**
- The histograms for each dataset confirm that \(Y\) follows a Chi-squared distribution with degrees of freedom equal to the dimensionality.
- The computed probabilities for \(Y \leq c^2\) match the theoretical Chi-squared CDF values, reinforcing the theoretical result.

## **Conclusion**
The study validates that the quadratic form \((x - \mu)^T \Sigma^{-1} (x - \mu)\) follows a Chi-squared distribution with degrees of freedom equal to the dimensionality of the multivariate normal distribution. This property is useful in statistical inference, anomaly detection, and hypothesis testing in multivariate analysis.

# Introduction

Bayesian classification is a statistical approach used for classifying data points based on probability distributions. This project implements a Bayesian classifier for a 2D dataset, where data points are classified into one of two classes using Multivariate Normal Distributions. The classification is performed using Bayes' theorem, which updates the probability of a class given the observed data.

# Methodology

**Define Class Distributions:**

Two classes are modeled as multivariate normal distributions with given mean vectors and covariance matrices.

**Compute Likelihoods:**

The likelihood of a data point belonging to each class is computed using the multivariate normal probability density function (PDF).

**Apply Bayes' Theorem:**

Posterior probabilities are calculated using:



Equal prior probabilities are assumed for both classes.

**Classify Data Points:**

A data point is assigned to the class with the higher posterior probability.

**Visualization:**

A scatter plot is generated, showing classified points in different colors based on their assigned class.



#Data

The dataset consists of 2D data points stored in a text file (File_Datapoints.txt).

Each row in the dataset contains:

x: The x-coordinate of the data point.

y: The y-coordinate of the data point.

Data is loaded using pandas for processing and classification.



```python
import scipy.stats as stats
import numpy as np
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
import pandas as pd

# Define the parameters for the two classes
mu1 = np.array([2, 3])
mu2 = np.array([-2, -3])
sigma1 = np.array([[1, 0.5], [0.5, 2]])
sigma2 = np.array([[2, -0.3], [-0.3, 1]])

# Prior probabilities (assuming equal priors for now)
prior1 = 0.5
prior2 = 0.5


def classify_point(x, y):
    x_vec = np.array([x, y])

    # Calculate likelihoods
    likelihood1 = multivariate_normal.pdf(x_vec, mean=mu1, cov=sigma1)
    likelihood2 = multivariate_normal.pdf(x_vec, mean=mu2, cov=sigma2)

    # Apply Bayes' theorem
    posterior1 = (likelihood1 * prior1) / (likelihood1 * prior1 + likelihood2 * prior2)
    posterior2 = (likelihood2 * prior2) / (likelihood1 * prior1 + likelihood2 * prior2)

    return 1 if posterior1 > posterior2 else 2


data = pd.read_csv('/content/File_Datapoints.txt', delimiter=' ', header=0)
x_values = data['x']
y_values = data['y']


# Classify data points and store results
classifications = [classify_point(x, y) for x, y in zip(x_values, y_values)]

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x_values[np.array(classifications) == 1], y_values[np.array(classifications) == 1], color='blue', label='Class 1')
plt.scatter(x_values[np.array(classifications) == 2], y_values[np.array(classifications) == 2], color='red', label='Class 2')

# Customize plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Classification of Data Points')
plt.legend()
plt.grid(True)
plt.show()
```

# Results

The data points were classified into two classes based on the Bayesian classification approach.

The classification results were visualized using a scatter plot, where:

**Class 1**
points are plotted in
**blue.**

**Class 2** points are plotted in **red.**

The classifier effectively separates data points based on their likelihoods and posterior probabilities.

# Conclusion

Bayesian classification provides an effective probabilistic approach for 2D data classification.

Using Multivariate Normal Distributions, we can model the spread and correlation of data.

The scatter plot visualization confirms that the classifier successfully differentiates between two classes.
